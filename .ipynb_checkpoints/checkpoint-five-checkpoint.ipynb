## Create Your Model

Use the two columns with the strongest correlation to create and train your model. Make sure to print out the summary and plot the column values and the line produced by the model.

```python
# Model for 2nd data set

# Creating X and y
X = passing_career_stats['Pass_Attempts_Per_Game']
y = passing_career_stats['Passer_Rating']
```

```python
# splitting the variables into training and testing 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)

# viewing the train dataset
X_train
```

```python
y_train
```

```python
# import statsmodel.api library from statsmodel package
import statsmodels.api as sm

# add constant to get an intercept
X_train_sm = sm.add_constant(X_train)

# fitting the regression line using 'Ordinary Least Square' OLS
lr = sm.OLS(y_train, X_train_sm).fit()

# printing parameters
lr.params
```

```python
# perform summary to list out all the different parameters of the regression line fitted 
lr.summary()
```

```python
# coefficient for Pass_Attempts_Per_Game is 2.13 & it's corresponding value is low, almost 0; meaning the coefficient
# is statistically significant [pvalue should always be less for the coefficient to be significant]
```

```python
# R-squared value is 0.478, meaning 47.8% of the Passer_Rating variance can be explained by the Pass_Attempts_Per_Game col
# using this line
```

```python
# Prob (F-statistic) has very low p-value, which means the model fit is statistically significant
```

```python

# simple linear regression formula
# y = c + mX
# y = c + m * Pass_Attempts_Per_Game
# Passer_Rating = 18.416 + 0.478 * Pass_Attempts_Per_Game
```

```python
# visualizing the regression line
plt.scatter(X_train, y_train)
plt.plot(X_train, 18.416+0.478*X_train, 'r')
plt.show()
```

## Error Terms

Finally, plot your error terms!

```python
# Error = Actual y value - y predicted value

# predicting y_value using training data of X using the 'predict' attr
y_train_pred = lr.predict(X_train_sm)

# creating residuals from the y_train data and predicted y_data
res = (y_train - y_train_pred)
```

```python
# plotting histogram using the residual values
fig = plt.figure()
sns.displot(res, bins=15, kde=True)
plt.title('Error Terms', fontsize=15)
plt.xlabel('y_train - y_train_pred', fontsize=15)
plt.show()
```

```python
# looking for any patterns in the residuals
plt.scatter(X_train, res)
plt.show()
```

```python
# residuals are not normally distributed and have a random pattern
```
